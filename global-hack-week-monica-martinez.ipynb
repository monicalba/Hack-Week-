#  Write a Sorting Method Task
Writing code to sort a list is another classic computer science problem. Using any language you’d like, sort a list!

```python
def insertion_sort(arr):
    n = len(arr)

    for i in range(1, n):
        key = arr[i]
        j = i - 1

        # Move elements of arr[0..i-1] that are greater than key to one position ahead of their current position
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
             # Swap the found minimum element with the first element

        arr[j + 1] = key

# Example list
my_list = [29, 43, 103, 305, 2484, 3948, 40384]

print("Original list:", my_list)
insertion_sort(my_list)
print("Sorted list:", my_list)
```

#  Build a Rock, Paper, Scissors Game Task
Write code to simulate a rock, paper, scissor game!



```python
import random
user_action = input("Enter a choice (rock, paper, scissors): ")
possible_actions = ["rock", "paper", "scissors"]
computer_action = random.choice(possible_actions)
print(f"\nYou chose {user_action}, computer chose {computer_action}.\n")
if user_action == computer_action:
    print(f"Both players selected {user_action}. It's a tie!")
elif user_action == "rock":
    if computer_action == "scissors":
        print("Rock smashes scissors! You win!")
    else:
        print("Paper covers rock! You lose.")
elif user_action == "paper":
    if computer_action == "rock":
        print("Paper covers rock! You win!")
    else:
        print("Scissors cuts paper! You lose.")
elif user_action == "scissors":
    if computer_action == "paper":
        print("Scissors cuts paper! You win!")
    else:
        print("Rock smashes scissors! You lose.")
        
```

#  Build a random number generator Task

```python
import random

class MyMersenneTwister:
    def __init__(self, seed=1):
        self.state = [seed]
        self.index = 0
        self.N = 624
        self.M = 397
        self.A = 0x9908B0DF
        self.UpperMask = 0x80000000
        self.LowerMask = 0x7FFFFFFF

        for i in range(1, self.N):
            prev = self.state[i - 1]
            next_val = (0x6C078965 * (prev ^ (prev >> 30)) + i) & 0xFFFFFFFF
            self.state.append(next_val)

    def extract_number(self):
        if self.index == 0:
            self.generate_numbers()

        y = self.state[self.index]
        y = y ^ (y >> 11)
        y = y ^ ((y << 7) & 0x9D2C5680)
        y = y ^ ((y << 15) & 0xEFC60000)
        y = y ^ (y >> 18)

        self.index = (self.index + 1) % self.N
        return y

    def generate_numbers(self):
        for i in range(self.N):
            y = (self.state[i] & self.UpperMask) + (self.state[(i + 1) % self.N] & self.LowerMask)
            self.state[i] = self.state[(i + self.M) % self.N] ^ (y >> 1)
            if y % 2 != 0:
                self.state[i] = self.state[i] ^ self.A

# Creating an instance of the random number generator
my_mersenne_twister = MyMersenneTwister()

# Generating random numbers
print("Random numbers:")
for _ in range(10):
    print(my_mersenne_twister.extract_number())

# Generating random numbers within a specific range
print("\nRandom numbers between 0 and 100:")
for _ in range(5):
    print(my_mersenne_twister.extract_number() % 101)

  
```

#  Create a Useless Hack
Create any silly, unnecessary, or useless hack!

```python
import random
import time

def multiplication_calculator():
    print("Welcome to the Multiplication Calculator!")

    while True:
        num1 = random.randint(1, 10)
        num2 = random.randint(1, 10)

        print(f"What is the product of {num1} and {num2}?")
        user_answer = input("Your Answer: ")

        time.sleep(1)  # Adding a delay for suspense

        actual_answer = num1 * num2
        if user_answer.isdigit() and int(user_answer) == actual_answer:
            print("Congratulations! You are correct.")
        else:
            print(f"Oops! The actual answer is {actual_answer}.")

        play_again = input("Do you want to try another multiplication? (yes/no): ")
        if play_again.lower() != 'yes':
            print("Thank you for using the Multiplication Calculator!")
            break

if __name__ == "__main__":
    multiplication_calculator()
```
